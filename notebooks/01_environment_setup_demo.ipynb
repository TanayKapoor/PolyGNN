```python
print("🧪 RDKit Check")
print("=" * 50)

try:
    from rdkit import Chem
    from rdkit import __version__ as rdkit_version
    print(f"✅ RDKit successfully imported")
    print(f"RDKit version: {rdkit_version}")

    # Test SMILES parsing
    test_smiles = [
        "CCO",        # Ethanol
        "c1ccccc1",   # Benzene
        "CC(=O)O",    # Acetic acid
        "CCc1ccccc1", # Ethylbenzene
    ]

    print("\nTesting SMILES parsing:")
    for smiles in test_smiles:
        mol = Chem.MolFromSmiles(smiles)
        if mol is not None:
            print(f"✅ '{smiles}' parsed successfully")
        else:
            print(f"❌ Failed to parse '{smiles}'")

    # Test molecular properties FOR ETHANOL
    print("\nMolecular properties:")
    mol_ethanol = Chem.MolFromSmiles("CCO")  # Use "CCO" for SMILES
    if mol_ethanol:
        # Add explicit hydrogens for correct atom/bond counting
        mol_ethanol_with_h = Chem.AddHs(mol_ethanol)
        
        # Calculate molecular properties using rdMolDescriptors
        from rdkit.Chem import rdMolDescriptors
        mol_formula = rdMolDescriptors.CalcMolFormula(mol_ethanol)
        print(f"  Molecule: {mol_formula} (Ethanol)")
        print(f"  Number of atoms (heavy only): {mol_ethanol.GetNumAtoms()}")
        print(f"  Number of atoms (with H): {mol_ethanol_with_h.GetNumAtoms()}")
        print(f"  Number of bonds (heavy only): {mol_ethanol.GetNumBonds()}")
        print(f"  Number of bonds (with H): {mol_ethanol_with_h.GetNumBonds()}")
        print(f"  Molecular weight: {rdMolDescriptors.CalcExactMolWt(mol_ethanol):.2f}")
        print("✅ Molecular property calculation works for ethanol")
    else:
        print("❌ Failed to create ethanol molecule from SMILES")

except ImportError as e:
    print(f"❌ Failed to import RDKit: {e}")
    print("Please install RDKit using: conda install rdkit -c conda-forge")
```

```python
print("🧠 Graph Neural Networks Check")
print("=" * 50)

try:
    import torch
    import torch.nn.functional as F
    from torch_geometric.nn import GCNConv
    from torch_geometric.data import Data
    
    # Create a simple graph (triangle)
    edge_index = torch.tensor([[0, 1, 1, 2, 2, 0],
                               [1, 0, 2, 1, 0, 2]], dtype=torch.long)
    
    # Node features (3 nodes, 4 features each)
    x = torch.randn(3, 4)
    
    # Create graph data object
    data = Data(x=x, edge_index=edge_index)
    
    print(f"Input graph:")
    print(f"  Nodes: {data.num_nodes}")
    print(f"  Edges: {data.num_edges}")
    print(f"  Input features shape: {data.x.shape}")
    
    # Create GCN layer
    gcn_layer = GCNConv(in_channels=4, out_channels=8)
    
    # Forward pass
    print(f"\nTesting GCN layer (4 -> 8 features)...")
    with torch.no_grad():
        output = gcn_layer(data.x, data.edge_index)
    
    print(f"Output shape: {output.shape}")
    print(f"Output mean: {output.mean().item():.4f}")
    print(f"Output std: {output.std().item():.4f}")
    
    # Test with activation
    print(f"\nTesting with ReLU activation...")
    with torch.no_grad():
        output_relu = F.relu(gcn_layer(data.x, data.edge_index))
    
    print(f"ReLU output shape: {output_relu.shape}")
    print(f"ReLU output mean: {output_relu.mean().item():.4f}")
    print(f"ReLU output std: {output_relu.std().item():.4f}")
    
    # Test multi-layer GCN
    print(f"\nTesting multi-layer GCN...")
    
    class SimpleGCN(torch.nn.Module):
        def __init__(self, input_dim, hidden_dim, output_dim):
            super(SimpleGCN, self).__init__()
            self.conv1 = GCNConv(input_dim, hidden_dim)
            self.conv2 = GCNConv(hidden_dim, output_dim)
            
        def forward(self, x, edge_index):
            x = F.relu(self.conv1(x, edge_index))
            x = self.conv2(x, edge_index)
            return x
    
    # Create and test multi-layer model
    model = SimpleGCN(input_dim=4, hidden_dim=16, output_dim=2)
    
    with torch.no_grad():
        final_output = model(data.x, data.edge_index)
    
    print(f"Multi-layer GCN output shape: {final_output.shape}")
    print(f"Multi-layer GCN output mean: {final_output.mean().item():.4f}")
    
    print("✅ Graph Neural Networks work correctly")
    
except ImportError as e:
    print(f"❌ Failed to import GNN modules: {e}")
except Exception as e:
    print(f"❌ Error in GNN operations: {e}")
```

```python
print("🔬 Polymer Processing Check")
print("=" * 50)

try:
    import torch
    from rdkit import Chem
    from rdkit.Chem import rdMolDescriptors
    from torch_geometric.data import Data
    
    # Test polymer-like molecules
    polymer_smiles = [
        "CCC(C)C",  # Branched alkane (polymer-like)
        "CCCCCCCCCC",  # Long chain alkane
        "c1ccc(CC)cc1",  # Aromatic with alkyl chain
        "CC(C)(C)C",  # Highly branched
        "CCCCCCCCCCCCCCCCCCCCC",  # Very long chain (polymer-like)
    ]
    
    print("Testing polymer-like molecules:")
    
    for i, smiles in enumerate(polymer_smiles):
        print(f"\n--- Molecule {i+1}: {smiles} ---")
        
        # Parse SMILES
        mol = Chem.MolFromSmiles(smiles)
        if mol is None:
            print(f"❌ Failed to parse SMILES: {smiles}")
            continue
        
        # Calculate molecular properties
        mw = rdMolDescriptors.CalcExactMolWt(mol)
        logp = rdMolDescriptors.CalcCrippenDescriptors(mol)[0]
        tpsa = rdMolDescriptors.CalcTPSA(mol)
        
        print(f"  ✅ Parsed successfully")
        print(f"  Molecular Weight: {mw:.2f}")
        print(f"  LogP: {logp:.2f}")
        print(f"  TPSA: {tpsa:.2f}")
        print(f"  Atoms: {mol.GetNumAtoms()}")
        print(f"  Bonds: {mol.GetNumBonds()}")
        
        # Convert to graph representation
        try:
            # Get atom features (simple: atomic number)
            atom_features = []
            for atom in mol.GetAtoms():
                atom_features.append([atom.GetAtomicNum()])
            
            # Get edge indices
            edge_indices = []
            for bond in mol.GetBonds():
                i = bond.GetBeginAtomIdx()
                j = bond.GetEndAtomIdx()
                edge_indices.extend([[i, j], [j, i]])  # Undirected graph
            
            if edge_indices:
                edge_index = torch.tensor(edge_indices, dtype=torch.long).t().contiguous()
                x = torch.tensor(atom_features, dtype=torch.float)
                
                # Create graph data object
                graph_data = Data(x=x, edge_index=edge_index)
                
                print(f"  Graph nodes: {graph_data.num_nodes}")
                print(f"  Graph edges: {graph_data.num_edges}")
                print(f"  Features shape: {graph_data.x.shape}")
                print(f"  ✅ Graph representation created")
            else:
                print(f"  ⚠️  No bonds found (single atom)")
                
        except Exception as e:
            print(f"  ❌ Error creating graph: {e}")
    
    print(f"\n✅ Polymer processing capabilities verified")
    
    # Test with a more complex polymer-like structure
    print(f"\n--- Testing Complex Polymer Structure ---")
    complex_polymer = "CC(C)(C)CC(C)(C)CC(C)(C)C"  # Branched polymer chain
    
    mol = Chem.MolFromSmiles(complex_polymer)
    if mol:
        print(f"Complex polymer: {complex_polymer}")
        print(f"  Atoms: {mol.GetNumAtoms()}")
        print(f"  Bonds: {mol.GetNumBonds()}")
        print(f"  MW: {rdMolDescriptors.CalcExactMolWt(mol):.2f}")
        print(f"  ✅ Complex polymer processing works")
    
except ImportError as e:
    print(f"❌ Failed to import required modules: {e}")
except Exception as e:
    print(f"❌ Error in polymer processing: {e}")
```
